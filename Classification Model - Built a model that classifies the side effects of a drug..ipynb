```python
#top 10 frequently used drugs most commonly used
top_10_drugs = data['Drug'].value_counts().head(10)

# Create a bar graph
plt.figure(figsize=(5,3))
top_10_drugs.plot(kind='bar', color='skyblue')
plt.xlabel('Drug')
plt.ylabel('Frequency')
plt.title('Top 10 Most Frequently Used Drugs')
plt.xticks(rotation=45)  # Rotate x-axis labels for readability
plt.show()
```

```python
#Most frequently taken drug in females
# Filter the DataFrame for Female gender
female_df = data[data['Sex'] == 'Female']

# Count the frequency of each drug and select the top 10
top_10_drugs = female_df['Drug'].value_counts().head(10)

# Create a bar graph
plt.figure(figsize=(5,3))
top_10_drugs.plot(kind='bar', color='skyblue')
plt.xlabel('Drug')
plt.ylabel('Frequency')
plt.title('Top 10 Drugs Most Frequently given for Females')
plt.xticks(rotation=45)  # Rotate x-axis labels for readability
plt.show()
```

```python
#drugs given exclusively for females only
# Group the filtered DataFrame by 'Drug' and count the number of occurrences
drug_counts = female_df['Drug'].value_counts()

# Select the drugs taken exclusively by females (occurs only once)
drugs_taken_exclusively_by_females = drug_counts[drug_counts == 1]

# Take the top 10 drugs
top_10_drugs = drugs_taken_exclusively_by_females.head(10)

# Create a bar graph
plt.figure(figsize=(5, 3))
top_10_drugs.plot(kind='bar', color='skyblue')
plt.xlabel('Drug')
plt.ylabel('Frequency')
plt.title(' 10 Drugs Taken Exclusively by Females')
plt.xticks(rotation=45)  # Rotate x-axis labels for readability
plt.show()
```

```python
#drugs given for males only
male_data = data[data['Sex'] == 'Male']
drug_counts_male = male_data['Drug'].value_counts()
# Group the filtered DataFrame by 'Drug' and count the number of occurrences
drug_counts_male = male_data['Drug'].value_counts()

# Select the drugs taken exclusively by females (occurs only once)
drugs_taken_exclusively_by_males = drug_counts_male[drug_counts_male == 1]

# Take the top 10 drugs
top_10_drugs = drugs_taken_exclusively_by_males.head(10)

# Create a bar graph
plt.figure(figsize=(5, 3))
top_10_drugs.plot(kind='bar', color='skyblue')
plt.xlabel('Drug')
plt.ylabel('Frequency')
plt.title('Top 10 Drugs Taken Exclusively by males')
plt.xticks(rotation=45)  # Rotate x-axis labels for readability
plt.show()
```

```python
#plotting the correlation 
# Calculate the correlation matrix
correlation_matrix = data.corr()

# Create a heatmap of the correlation matrix
plt.figure(figsize=(5, 4))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f", square=True)
plt.title('Correlation Matrix Heatmap')
plt.show()
```

```python
# Define the desired order of age categories
age_order = ['0-2', '3-6', '7-12', '13-18', '19-24', '25-34', '35-44', '45-54', '55-64', '65-74', '75 or over']

# Convert 'Age' column to categorical with desired order
df['Age'] = pd.Categorical(df['Age'], categories=age_order, ordered=True)

# Map age categories to numeric values
age_numeric = df['Age'].map({category: i for i, category in enumerate(age_order)})

plt.figure(figsize=(8, 5))

# Create histogram
plt.hist(age_numeric, bins=len(age_order), color='blue',edgecolor='black')

# Set x-axis tick labels
plt.xticks(range(len(age_order)), age_order,rotation=90)

# Set labels and title
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.title('Age Distribution',color='black')

# Show the plot
plt.show() 
```

```python
for col in ['EaseofUse','Satisfaction']:
    plt.figure(figsize=(10, 6))
    sns.countplot(data=data, x=col)
    plt.title(f'Distribution of {col}')
    plt.xlabel(col)
    plt.ylabel('Count')
    plt.xticks(rotation = 90)
    plt.xticks(rotation=90)
    plt.show()
```

```python

# Calculate the top 10 drugs for each age group
top_10_drugs_by_age_group = data.groupby('Age')['Drug'].value_counts().groupby(level=0, group_keys=False).nlargest(10).reset_index(name='Count')

# Get unique age groups
age_groups = top_10_drugs_by_age_group['Age'].unique()

# Create a separate bar plot for each age group
for age_group in age_groups:
    age_group_data = top_10_drugs_by_age_group[top_10_drugs_by_age_group['Age'] == age_group]

    plt.figure(figsize=(8, 4))
    plt.bar(age_group_data['Drug'], age_group_data['Count'], color = 'g')
    plt.xlabel('Drug')
    plt.ylabel('Count')
    plt.title(f'Top 10 Drugs for Age Group: {age_group}')
    plt.xticks(rotation=45, ha='right')
    plt.tight_layout()
    plt.show()
```
